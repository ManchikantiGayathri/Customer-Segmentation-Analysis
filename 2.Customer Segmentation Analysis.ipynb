# 3.Customer Segmentation (K-Means Clustering) ðŸ§‘â€ðŸ¤â€ðŸ§‘

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is already loaded and cleaned, and 'TotalPurchases' is created

print("\n--- Customer Segmentation using K-Means Clustering ---")

# Feature Selection for Clustering
# Select features relevant for customer behavior and demographics
features_for_clustering = data[[
    'Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts',
    'MntFishProducts', 'MntSweetProducts', 'MntGoldProds',
    'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
    'NumStorePurchases', 'NumWebVisitsMonth', 'Age', 'MntTotal', 'TotalPurchases'
]].copy() # Use .copy() to avoid SettingWithCopyWarning

# Handle any potential remaining NaNs in selected features before scaling (e.g., if Age or other derived columns were missing)
# For this dataset, Income was handled, and others are typically complete.
# If any NaNs appear here, consider median/mean imputation or dropping rows.
if features_for_clustering.isnull().any().any():
    print("Warning: Missing values found in features selected for clustering. Imputing with median.")
    for col in features_for_clustering.columns:
        if features_for_clustering[col].isnull().any():
            features_for_clustering[col].fillna(features_for_clustering[col].median(), inplace=True)


# Feature Scaling: K-Means is sensitive to the scale of features.
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features_for_clustering)
scaled_data = pd.DataFrame(scaled_features, columns=features_for_clustering.columns)

print("\nScaled Features Head (first 5 rows):")
print(scaled_data.head())

# Determine Optimal Number of Clusters (Elbow Method)
sse = {} # Sum of squared distances
for k in range(1, 11): # Test k from 1 to 10
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # n_init for robust centroid initialization
    kmeans.fit(scaled_data)
    sse[k] = kmeans.inertia_ # Inertia is the SSE
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols_for_clustering = ['Income', 'MntTotal', 'In_relationship']
data_scaled = data.copy()
data_scaled[cols_for_clustering] = scaler.fit_transform(data[cols_for_clustering])
data_scaled[cols_for_clustering].describe()
```

### Principal Component Analysis (PCA)

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components = 2)
pca_res = pca.fit_transform(data_scaled[cols_for_clustering])
data_scaled['pc1'] = pca_res[:,0]
data_scaled['pc2'] = pca_res[:,1]
```

### Elbow method

```python
X = data_scaled[cols_for_clustering]
inertia_list = []
for K in range(2,10):
    inertia = KMeans(n_clusters=K, random_state=7).fit(X).inertia_
    inertia_list.append(inertia)
```

```python
plt.figure(figsize=[7,5])
plt.plot(range(2,10), inertia_list, color=(54 / 255, 113 / 255, 130 / 255))
plt.title("Inertia vs. Number of Clusters")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.show()
```

### Silhouette score analysis

```python
from sklearn.metrics import silhouette_score
silhouette_list = []
for K in range(2,10):
    model = KMeans(n_clusters = K, random_state=7)
    clusters = model.fit_predict(X)
    s_avg = silhouette_score(X, clusters)
    silhouette_list.append(s_avg)

plt.figure(figsize=[7,5])
plt.plot(range(2,10), silhouette_list, color=(54 / 255, 113 / 255, 130 / 255))
plt.title("Silhouette Score vs. Number of Clusters")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.show()
```

```python
model = KMeans(n_clusters=4, random_state = 7)
model.fit(data_scaled[cols_for_clustering])
data_scaled['Cluster'] = model.predict(data_scaled[cols_for_clustering])
```

#### New feature: Marital

```python
def get_marital_status(row):
    if row['marital_Divorced'] == 1:
        return 'Divorced'
    elif row['marital_Married'] == 1:
        return 'Married'
    elif row['marital_Single'] == 1:
        return 'Single'
    elif row['marital_Together'] == 1:
        return 'Together'
    elif row['marital_Widow'] == 1:
        return 'Widow'
    else:
        return 'Unknown'
data['Marital'] = data.apply(get_marital_status, axis=1)
```

## DBSCAN Clustering for Customer Segmentation ðŸ§‘â€ðŸ¤â€ðŸ§‘

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA # For visualization

print("\n--- Customer Segmentation using DBSCAN ---")

# Step 1: Determine optimal eps and min_samples
eps_value = 0.5 
min_samples_value = 5

# Step 2: Apply DBSCAN
dbscan = DBSCAN(eps=eps_value, min_samples=min_samples_value)
data['DBSCAN_Cluster'] = dbscan.fit_predict(scaled_data)
```

```python
# Clusters are labeled from 0 to n-1, and -1 indicates noise points
print(f"DBSCAN applied with eps={eps_value} and min_samples={min_samples_value}.")
print("Distribution of customers per DBSCAN cluster (including noise -1):")
print(data['DBSCAN_Cluster'].value_counts().sort_index())
```

```python
# Step 3: Analyze and Visualize DBSCAN Clusters
meaningful_clusters_data = data[data['DBSCAN_Cluster'] != -1]
if not meaningful_clusters_data.empty:
    dbscan_cluster_means = meaningful_clusters_data.groupby('DBSCAN_Cluster')[features_for_clustering.columns].mean()
    print("\nMean of Features per DBSCAN Cluster (Original Scale, excluding noise):")
    print(dbscan_cluster_means)
else:
    print("\nNo meaningful clusters found by DBSCAN (all points might be noise or not enough points per cluster).")
```

```python
# Visualization: Use PCA to reduce dimensions for 2D scatter plot
pca = PCA(n_components=2)
principal_components_dbscan = pca.fit_transform(scaled_data)
pca_dbscan_data = pd.DataFrame(data=principal_components_dbscan, columns=['PC1', 'PC2'])
pca_dbscan_data['DBSCAN_Cluster'] = data['DBSCAN_Cluster']

plt.figure(figsize=(10, 8))
sns.scatterplot(x='PC1', y='PC2', hue='DBSCAN_Cluster', data=pca_dbscan_data,
                palette='tab10', s=100, alpha=0.7,
                legend='full' # Show legend for all clusters and noise
               )
plt.title(f'Customer Segments (DBSCAN with eps={eps_value}, min_samples={min_samples_value})')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()

print("\n--- DBSCAN Clustering Complete ---")
```

## Types of Errors for Model Performance (Classification) ðŸ“‰

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

print("--- Demonstrating Types of Errors in Classification ---")

# Step 1: Create a synthetic dataset for a binary classification problem
# Imagine we're predicting if a customer will churn (1) or not (0)
np.random.seed(42) # for reproducibility

# Features (e.g., spending, recency, num_complaints)
X = np.random.rand(200, 3) * 100

# Target variable (0: No Churn, 1: Churn)
# Let's make it somewhat dependent on features
y = (X[:, 0] + X[:, 1] / 2 - X[:, 2] / 3 + np.random.normal(0, 10, 200) > 80).astype(int)

# Convert to DataFrame (optional, but good practice)
data_classification = pd.DataFrame(X, columns=['Feature1', 'Feature2', 'Feature3'])
data_classification['Churn'] = y

print("\nSynthetic Classification Dataset Head:")
print(data_classification.head())

# Step 2: Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"\nTraining set size: {len(X_train)} samples")
print(f"Testing set size: {len(X_test)} samples")
print(f"Actual Churn in Test Set (1s): {np.sum(y_test == 1)}")
print(f"Actual No Churn in Test Set (0s): {np.sum(y_test == 0)}")


# Step 3: Train a simple classification model (Logistic Regression)
model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)

# Step 4: Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Step 5: Generate the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)

# Visualize the Confusion Matrix
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted 0 (No Churn)', 'Predicted 1 (Churn)'],
            yticklabels=['Actual 0 (No Churn)', 'Actual 1 (Churn)'])
plt.title('Confusion Matrix')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.show()

# Step 6: Extract Type I and Type II Errors (and other metrics)
# A confusion matrix is typically structured as:
# [[TN, FP],
#  [FN, TP]]

TN, FP, FN, TP = cm.ravel()

print(f"\nTrue Negatives (TN): {TN} (Correctly predicted No Churn)")
print(f"False Positives (FP - Type I Error): {FP} (Incorrectly predicted Churn when actual was No Churn)")
print(f"False Negatives (FN - Type II Error): {FN} (Incorrectly predicted No Churn when actual was Churn)")
print(f"True Positives (TP): {TP} (Correctly predicted Churn)")

# Step 7: Print Classification Report for more detailed metrics
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['No Churn', 'Churn']))

print("\n--- End of Error Types Demonstration ---")
```
